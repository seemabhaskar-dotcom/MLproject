#### 4.3 Multivariate analysis using pieplot

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example: Binning math_score into categories
df['math_score_group'] = pd.cut(df['math_score'],
                                bins=[0, 50, 75, 100],
                                labels=['Low', 'Average', 'High'])

# Define the categorical columns and titles
columns_info = {
    'gender': 'Gender',
    'race_ethnicity': 'Race/Ethnicity',
    'lunch': 'Lunch',
    'test_preparation_course': 'Test Preparation Course',
    'parental_level_of_education': 'Parental Level of Education',
    'math_score_group': 'Math Score Group'
}

# Set up the plot grid (2 rows x 3 columns)
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 12))
axes = axes.flatten()  # Flatten for easy iteration

# Define some default colors
colors = ['red', 'green', 'blue', 'cyan', 'orange', 'gray', 'purple']

# Generate pie charts
for i, (col, title) in enumerate(columns_info.items()):
    sizes = df[col].value_counts()
    labels = sizes.index.tolist()
    axes[i].pie(sizes,
                labels=labels,
                colors=colors[:len(labels)],
                autopct='%.2f%%',
                startangle=140)
    axes[i].set_title(title, fontsize=14)
    axes[i].axis('equal')  # Circle
    axes[i].axis('off')    # Clean

# Remove any extra subplot if columns < subplots (optional safeguard)
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

#####  Insights
- Number of Male and Female students is almost equal
- Number students are greatest in Group C
- Number of students who have standard lunch are greater
- Number of students who have not enrolled in any test preparation course is greater
- Number of students whose parental education is "Some College" is greater followed closely by "Associate's Degree"

#### 4.4.5 TEST PREPARATION COURSE COLUMN 
- Which type of lunch is most common amoung students ?
- Is Test prepration course has any impact on student's performance ?

#### BIVARIATE ANALYSIS ( Is Test prepration course has any impact on student's performance ? )

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for consistent appearance
sns.set(style="whitegrid", palette="Set2", font_scale=1.1)

# Create subplots
plt.figure(figsize=(15, 15))

# Math Score
plt.subplot(2, 2, 1)
sns.barplot(data=df, x='lunch', y='math_score', hue='test_preparation_course', ci=None)
plt.title("Math Score vs Lunch by Test Prep", fontweight='bold')
plt.ylabel("Math Score")
plt.xlabel("Lunch Type")
plt.legend(title="Test Preparation")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Reading Score
plt.subplot(2, 2, 2)
sns.barplot(data=df, x='lunch', y='reading_score', hue='test_preparation_course', ci=None)
plt.title("Reading Score vs Lunch by Test Prep", fontweight='bold')
plt.ylabel("Reading Score")
plt.xlabel("Lunch Type")
plt.legend(title="Test Preparation")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Writing Score
plt.subplot(2, 2, 3)
sns.barplot(data=df, x='lunch', y='writing_score', hue='test_preparation_course', ci=None)
plt.title("Writing Score vs Lunch by Test Prep", fontweight='bold')
plt.ylabel("Writing Score")
plt.xlabel("Lunch Type")
plt.legend(title="Test Preparation")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust spacing and layout
plt.suptitle("Impact of Test Preparation Course on Student Performance by Lunch Type", fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

#### Insights
